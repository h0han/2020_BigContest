```python
# 데이터 셋에 피장타율, 방어율, 삼진/볼넷, 피출루율, 피OPS, 잔루처리율 column을 추가합니다.
columns = ['피장타율', '방어율', '삼진/볼넷', '피출루율', '피OPS', '잔루처리율']
for column in columns:
    df_full[column] = np.nan
```

```python
df_check = pd.read_csv("20200927기준_상대팀별누적투수데이터_선발(수정).csv", encoding='cp949')
```

```python
df_check
```

```python
# 선발 투수 데이터를 형성하는 과정입니다.
def match_data(df_pit, df_check):
    for i in range(len(df_pit['P_ID'])):
        player = df_pit['P_ID'][i]
        vsteam = df_pit['VS_T_ID'][i] # 플레이어 ID와 상대 팀 ID를 입력받습니다.
        temp_player = df_check[ df_check['P_ID']==player ]
        temp_vsteam = temp_player[ temp_player['VS_T_ID']==vsteam ] # 입력받은 데이터를 토대로 기존의 데이터를 불러옵니다.
        temp_arr = temp_vsteam.loc[:, ['피장타율', '방어율', '삼진/볼넷', '피출루율', '피OPS', '잔루처리율']].values
        columns = ['피장타율', '방어율', '삼진/볼넷', '피출루율', '피OPS', '잔루처리율']
        
        if len(temp_arr)==0: # 해당 선발 투수가 해당 상대팀을 상대한 전적이 없을 경우 빈칸으로 채웁니다.
            for j in range(len(columns)):
                column = columns[j]
                df_pit[column][i] = np.nan
            continue
        
        else:             # 선발 투수의 데이터를 각 column에 채웁니다.
            for j in range(len(columns)):
                column = columns[j]
                value = temp_arr[0][j]
                df_pit[column][i] = value
```

```python
match_data(df_full, df_check)
```
