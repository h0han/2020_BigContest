```python
df_total = pd.read_csv('Predict_input_data_선발투수.csv', encoding = 'cp949', engine = 'python')
df_bat = pd.read_csv('20200927기준_상대팀별누적타자데이터.csv', encoding = 'cp949', engine = 'python')
```

```python
df_total.drop(['P_ID', '피장타율', '방어율', '삼진/볼넷', '피출루율', '피OPS', '잔루처리율'], axis = 1, inplace = True)
```

```python
go = []
for i in range(len(df_total)):
    team = df_total['T_ID'][i] 
    vsteam = df_total['VS_T_ID'][i] 
    go.append(df_bat[df_bat['T_ID']==vsteam][df_bat['VS_T_ID']==team][list(df_bat.columns)[2:]].values[0])
```

```python
mod = pd.DataFrame(data=go, columns=list(df_bat.columns)[2:] )
```

```python
df = pd.concat([df_total[list(df_total.columns)[1:4]], mod],axis=1)
df
```

```python
df.to_csv('Predict_input_data_타자.csv', sep = ',', na_rep = 'NaN', encoding = 'ms949')
```
