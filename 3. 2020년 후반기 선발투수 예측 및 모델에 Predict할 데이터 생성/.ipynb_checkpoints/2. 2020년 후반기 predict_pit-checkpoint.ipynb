# Make Original Data

원본 기록은 http://www.kbreport.com/ 에서 수집하였음.<br>
홈페이지 내 선수기록 - 투구기록 - 세부기록 메뉴에서 상대팀별 기록 수집.<br>
수집 기준 날짜는 2020-09-27 23:16

```python
teamlist = ['LG', 'HT', 'KT', 'OB', 'WO', 'SS', 'LT', 'NC', 'HH', 'SK']
```

```python
df = pd.DataFrame()
for team in teamlist:
    vsteamlist = teamlist.copy()
    for vsteam in vsteamlist:
        if team!=vsteam:
            temp = pd.read_html('2020 상대팀대상 누적투수데이터/{}/{}_{}.html'.format(team, team, vsteam))
            temp = temp[0]
            df = pd.concat([df, temp])
```

```python
del df['#']
replace = {'두산':'OB', '삼성':'SS', 'KIA':'HT', 'Hero':'WO', '한화':'HH', '롯데':'LT', '이용찬*':'이용찬'}
df = df.replace(replace)
df.rename(columns = {'팀명':'T_ID', '상대':'VS_T_ID', '선수명':'P_ID', '삼/볼':'삼진/볼넷', 'ERA':'방어율', 'LOB%':'잔루처리율'}, inplace = True)
df = df[['T_ID', 'VS_T_ID', 'P_ID', '피장타율', '방어율', '삼진/볼넷', '피출루율', '피OPS', '잔루처리율']]

df
```

# P_ID 코드로 치환

```python
code_list = pd.read_csv('datasets/2020빅콘테스트_스포츠투아이_제공데이터_선수_2020.csv', encoding='cp949')
code_list
```

```python
name_list = list(code_list['NAME'])
pcode = list(code_list['PCODE'])
code_dict = dict()

for idx in range(len(name_list)):
    code_dict[name_list[idx]] = pcode[idx]
```

```python
df = df.replace(code_dict)
df
```

```python
df.to_csv('20200927기준_상대팀별누적투수데이터_선발.csv', index= False, encoding='cp949')
```
