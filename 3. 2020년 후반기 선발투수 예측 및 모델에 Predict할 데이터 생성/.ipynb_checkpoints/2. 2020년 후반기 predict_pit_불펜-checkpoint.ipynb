# Make Original Data

원본 기록은 http://www.kbreport.com/ 에서 수집하였음.<br>
홈페이지 내 선수기록 - 투구기록 - 세부기록 메뉴에서 상대팀별 기록 수집.<br>
수집 기준 날짜는 2020-09-27 23:16

```python
df = pd.read_excel('불펜데이터_크롤링.xlsx', encoding='cp949')
df
```

```python
replace = {'두산':'OB', '삼성':'SS', 'KIA':'HT', '키움':'WO', '한화':'HH', '롯데':'LT'}
df = df.replace(replace)
df.rename(columns = {'이름':'T_ID', '팀':'VS_T_ID', '방어율':'방어율_', '피출루율':'피출루율_', '피OPS':'피OPS_', '삼진/볼넷':'삼진/볼넷_', '피장타율':'피장타율_', '잔루처리율':'잔루처리율_'}, inplace = True)

df
```

```python
df.to_csv('20200927기준_상대팀별누적투수데이터_불펜.csv', index= False, encoding='cp949')
```
