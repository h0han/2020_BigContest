# Make Original Data

원본 기록은 http://www.kbreport.com/ 에서 수집하였음.<br>
홈페이지 내 팀기록 - 타격기록 메뉴에서 상대팀별 기록 수집.<br>
수집 기준 날짜는 2020-09-27 23:16

```python
teamlist = ['LG', 'HT', 'KT', 'OB', 'WO', 'SS', 'LT', 'NC', 'HH', 'SK']
```

```python
df = pd.DataFrame()
for team in teamlist:
    vsteamlist = teamlist.copy()
    vsteamlist.pop(vsteamlist.index(team))
    for vsteam in vsteamlist:
        temp = pd.read_html('2020 상대팀대상 누적타자데이터/{}_{}_.html'.format(team, vsteam), encoding='cp949')
        temp = temp[0]
        temp = temp.iloc[0,]
        temp = pd.DataFrame(temp).T
        df = pd.concat([df, temp])
```

```python
del df['#']; del df['WAR']; del df['기대승률']
replace = {'두산':'OB', '삼성':'SS', 'KIA':'HT', 'Hero':'WO', '한화':'HH', '롯데':'LT'}
df = df.replace(replace)
df.rename(columns = {'팀명':'T_ID', '상대':'VS_T_ID', 'OPS':'cOPS', '출루율':'cOBP', '장타율':'cSLG', 'BABIP':'cBABIP'}, inplace = True)
df['KK/BB'] = df['삼진%']/df['볼넷%']

drop_list = ['경기', '승', '무', '패', '승률', 'R/G', '득점', '홈런', '도루', '타율', '볼넷%', '삼진%']
df = df.drop(drop_list, axis=1)
df = df.reset_index()
del df['index']

df = df[['T_ID', 'VS_T_ID', 'cOBP', 'cSLG', 'cOPS', 'cBABIP', 'KK/BB', 'wOBA']]
```
